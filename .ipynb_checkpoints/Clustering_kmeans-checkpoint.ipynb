```python
import pandas as pd
import numpy as np
from operator import itemgetter
import math

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
def distancia_euc(lat1, lon1, lat2, lon2):
    p = math.pi/180
    a = 0.5 - math.cos((lat2-lat1)*p)/2 + math.cos(lat1*p) * math.cos(lat2*p) * (1-math.cos((lon2-lon1)*p))/2
    d = 12742 * math.asin(math.sqrt(a))
    return  d
```

```python
def kmeans_clusters_radios(data_delitos_original,n_clusters):
    
    #Crear copia de base de datos y obtener coordenadas
    data_delitos_kmeans = data_delitos_original.copy()
    X = data_delitos_kmeans.loc[:,['longitud','latitud']]
    
    #Instanciar modelo y alimentar datos
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    y_kmeans = kmeans.fit_predict(X)
    
    #Añadir labels a dataframe
    data_delitos_kmeans['kmeans_labels'] = kmeans.labels_
    
    #Obtener centroides y radios.
    centroide_cluster = dict()
    max_vector_cluster = dict()

    for cluster in list(set(kmeans.labels_)):

        centroide_cluster[cluster]=list(zip(kmeans.cluster_centers_[:, 0],kmeans.cluster_centers_[:,1]))[cluster]

        vector_list = []
        for i in zip(data_delitos_kmeans[data_delitos_kmeans['kmeans_labels']==cluster].loc[:,'longitud'],data_delitos_kmeans[data_delitos_kmeans['kmeans_labels']==cluster].loc[:,'latitud']):
                vector_list.append((i,np.linalg.norm(np.subtract(i,centroide_cluster[cluster]))))
        max_vector_cluster[cluster] = max(vector_list, key=itemgetter(1))[0]
    
    #Obtener max radios
    radio_clusters = dict()
    radio_clusters = [distancia_euc(centroide_cluster[i][1],centroide_cluster[i][0],max_vector_cluster[i][1],max_vector_cluster[i][0]) for i in set(kmeans.labels_)]
    
    
    return len(np.unique(kmeans.labels_)),max(radio_clusters), data_delitos_kmeans
```

```python
#Ejecutar la clusterización para el rango de número clústeres elegido.
resultados = dict()
for i in range(50,100,10):
    clusters, radio = kmeans_clusters_radios(data_delitos,i)
    resultados[clusters]=radio
```
