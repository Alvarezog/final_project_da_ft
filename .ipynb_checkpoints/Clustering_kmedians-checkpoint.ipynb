```python
def distancia_euc(lat1, lon1, lat2, lon2):
    p = math.pi/180
    a = 0.5 - math.cos((lat2-lat1)*p)/2 + math.cos(lat1*p) * math.cos(lat2*p) * (1-math.cos((lon2-lon1)*p))/2
    d = 12742 * math.asin(math.sqrt(a))
    return  d
```

```python
def kmedians_clusters_radios(X,n_clusters,centroides):
    
    #Crear copia de base de datos y obtener coordenadas
    data_delitos_kmedians = np.array([X['longitud'],X['latitud']])
    data_delitos_kmedians = data_delitos_kmedians.transpose()
    data_delitos_kmedians = data_delitos_kmedians.tolist()

    #Crear lista con medias iniciales
    centros = centroides[str(n_clusters)]
    centros_lista = [centros[str(i)] for i in range(n_clusters)]
    
    #Instanciar modelo y alimentar datos
    kmedians_inst = kmedians(data_delitos_kmedians,centros_lista)
    kmedians_inst.process()
    y_kmedians = kmedians_inst.predict(data_delitos_kmedians)
    medians = kmedians_inst.get_medians()
    
    #Añadir labels a dataframe
    X['kmedians_labels'] = y_kmedians
    
    #Obtener centroides y radios.
    centroide_cluster = dict()
    max_vector_cluster = dict()

    for cluster in list(set(y_kmedians)):

        centroide_cluster[cluster]= medians[cluster]

        vector_list = []
        for i in zip(X[X['kmedians_labels']==cluster].loc[:,'longitud'],X[X['kmedians_labels']==cluster].loc[:,'latitud']):
                vector_list.append((i,np.linalg.norm(np.subtract(i,centroide_cluster[cluster]))))
        max_vector_cluster[cluster] = max(vector_list, key=itemgetter(1))[0]
    
    #Obtener max radios
    radio_clusters = dict()
    radio_clusters = [distancia_euc(centroide_cluster[i][1],centroide_cluster[i][0],max_vector_cluster[i][1],max_vector_cluster[i][0]) for i in set(y_kmedians)]
    
    
    return len(np.unique(y_kmedians)),max(radio_clusters), X, radio_clusters, centroide_cluster
```

```python
# Importa base de datos
data_delitos = pd.read_csv('/content/drive/MyDrive/Datasets/data_limpia.csv')
data_delitos.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# Importa los centros
centroides_json = open('/content/drive/MyDrive/Proyecto Final IronHack/Primer experimento kmeans/centroides_clusters_kmeans.json')
centroides_apriori = json.load(centroides_json)
```

```python
#Ejecutar la clusterización para el rango de número clústeres elegido.
resultados = dict()
radios = dict()
centroides = dict()
for i in range(120,210,10):
    clusters, radio, data_delitos_kmedians, radio_clusters, centroide_cluster = kmedians_clusters_radios(data_delitos,i,centroides_apriori)
    resultados[clusters]=radio
    data_delitos_kmedians.to_csv(f'data_delitos_kmed{i}.csv')
    radios[clusters] = radio_clusters
    centroides[clusters] = centroide_cluster
```

```python
#Calcular el radio máximo al 95% de significancia.
resultados = dict()
for n_k in radios:
    perc_95 = np.percentile(radios[n_k],95)
    radio_max = 0
    for radio in radios[n_k]:
        if radio <= perc_95 and radio > radio_max:
            radio_max = radio
    resultados[n_k]= radio_max
```
