```python
centroides = [centroides_ideal['150'][i] for i in centroides_ideal['150']]
```

```python
centroides_inverso = [[i[1],i[0]] for i in centroides]
```

```python
centroides_string = [str(coord[0]) + ", " + str(coord[1]) for coord in centroides_inverso]
```

```python
responses = []
palabra_clave = "policia"
API_KEY = "Insert google maso api key here"
for centroide in centroides_string:
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={}&radius=2000&keyword={}&key={}".format(centroide,palabra_clave, API_KEY)
    payload={}
    headers = {}

    response = requests.request("GET", url, headers=headers, data=payload)
    responses.append(response)
```
